```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(start=0,stop=2*np.pi,step=np.pi/20)
y=np.sin(x)
```

```python
plt.figure(figsize=(8,6))
plt.title('Discrete Sinusoidal Signal')
plt.scatter(x=x,y=y);
```

```python
x2=np.arange(start=-25,stop=25,step=2)
y2=np.exp(x2)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.title('Discrete Exponential Signal')
plt.scatter(x=x2,y=y2)
plt.subplot(1,2,2)
plt.title('Discrete Sinusoidal Signal')
plt.scatter(x=x,y=y);
```

```python
y3=np.sin(x-np.pi/2)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.title('Discrete Sinusoidal Signal(w/o phase shift)')
plt.scatter(x=x,y=y)
plt.subplot(1,2,2)
plt.title('Discrete Sinusoidal Signal(with phase shift)')
plt.scatter(x=x,y=y3);
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y,label='Sin wave')
plt.plot(x,y3,label='Cosine wave')
plt.title('Periodic Continous Waves')
plt.legend();
```
